```python
!pip install scikit-fuzzy ipywidgets

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
from ipywidgets import interact

# -------------------------
# ВХОДНИ ПРЕМЕНЛИВИ
# -------------------------
income = ctrl.Antecedent(np.arange(0, 10001, 1), 'income')
debt = ctrl.Antecedent(np.arange(0, 10001, 1), 'debt')
history = ctrl.Antecedent(np.arange(0, 11, 1), 'history')
age = ctrl.Antecedent(np.arange(18, 71, 1), 'age')

risk = ctrl.Consequent(np.arange(0, 101, 1), 'risk')

# -------------------------
# MEMBERSHIP ФУНКЦИИ
# -------------------------

income['low'] = fuzz.trimf(income.universe, [0, 0, 4000])
income['medium'] = fuzz.trimf(income.universe, [2000, 5000, 8000])
income['high'] = fuzz.trimf(income.universe, [6000, 10000, 10000])

debt['low'] = fuzz.trimf(debt.universe, [0, 0, 3000])
debt['medium'] = fuzz.trimf(debt.universe, [2000, 5000, 8000])
debt['high'] = fuzz.trimf(debt.universe, [6000, 10000, 10000])

history['bad'] = fuzz.trimf(history.universe, [0, 0, 4])
history['average'] = fuzz.trimf(history.universe, [3, 5, 7])
history['good'] = fuzz.trimf(history.universe, [6, 10, 10])

age['young'] = fuzz.trimf(age.universe, [18, 18, 30])
age['middle'] = fuzz.trimf(age.universe, [25, 40, 55])
age['old'] = fuzz.trimf(age.universe, [50, 70, 70])

risk['low'] = fuzz.trimf(risk.universe, [0, 0, 40])
risk['medium'] = fuzz.trimf(risk.universe, [30, 50, 70])
risk['high'] = fuzz.trimf(risk.universe, [60, 100, 100])

# -------------------------
# ПРАВИЛА
# -------------------------
rules = [
    ctrl.Rule(income['high'] & debt['low'] & history['good'], risk['low']),
    ctrl.Rule(income['medium'] & debt['medium'], risk['medium']),
    ctrl.Rule(debt['high'] | history['bad'], risk['high']),
    ctrl.Rule(income['low'] & debt['high'], risk['high']),
    ctrl.Rule(income['high'] & history['bad'], risk['medium']),
    ctrl.Rule(age['young'] & debt['high'], risk['high']),
    ctrl.Rule(age['old'] & history['good'], risk['low']),
    ctrl.Rule(income['medium'] & history['good'], risk['low']),
    ctrl.Rule(income['low'] & history['good'], risk['medium']),
    ctrl.Rule(age['middle'] & debt['medium'], risk['medium']),
    ctrl.Rule(age['young'] & income['low'], risk['high']),
]

# -------------------------
# СИСТЕМА
# -------------------------
system = ctrl.ControlSystem(rules)
sim = ctrl.ControlSystemSimulation(system)

# -------------------------
# ФУНКЦИЯ ЗА ОЦЕНКА
# -------------------------
def evaluate(income_val, debt_val, history_val, age_val):
    sim.input['income'] = income_val
    sim.input['debt'] = debt_val
    sim.input['history'] = history_val
    sim.input['age'] = age_val

    sim.compute()

    result = sim.output['risk']
    print(f"\nRisk score: {result:.2f}")

    if result < 40:
        print("Категория: НИСЪК РИСК")
    elif result < 70:
        print("Категория: СРЕДЕН РИСК")
    else:
        print("Категория: ВИСОК РИСК")

    risk.view(sim=sim)
    plt.show()

# -------------------------
# ТЕСТОВИ СЦЕНАРИИ
# -------------------------
print("=== ТЕСТ 1 ===")
evaluate(8000, 2000, 8, 45)

print("=== ТЕСТ 2 ===")
evaluate(3000, 7000, 3, 25)

print("=== ТЕСТ 3 ===")
evaluate(5000, 4000, 6, 60)

# -------------------------
# ИНТЕРАКТИВЕН РЕЖИМ
# -------------------------
def interactive_eval(income, debt, history, age):
    evaluate(income, debt, history, age)

interact(
    interactive_eval,
    income=(0, 10000, 500),
    debt=(0, 10000, 500),
    history=(0, 10, 1),
    age=(18, 70, 1)
)
```
